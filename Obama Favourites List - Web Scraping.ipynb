```python
#Create function to get all movie data
df=pd.DataFrame()
def scrapemovieinfo(URL):
    global df
    
    #Get movie title
    movieurl=URL
    get_text=requests.get(movieurl).content
    
    #First check URl works
    try:
        movietitle=(BeautifulSoup(get_text)).find('h1',class_='scoreboard__title').text
    
        #Get movieinfo
        movieinfo=[]
        rowlabel=[]
        for i in range(len(BeautifulSoup(get_text,'html.parser').find_all\
                   ("div",class_='meta-value'))):
            movieinfo.append((BeautifulSoup(get_text,'html.parser').find_all\
                   ("div",class_='meta-value')[i].text).replace('\n','').strip())
            rowlabel.append((BeautifulSoup(get_text,'html.parser').find_all\
                   ("div",class_='meta-label subtle')[i].text).replace('\n','').strip())

        #Get moviescores
        audiencescore=BeautifulSoup(get_text,'html.parser').find("score-board")['audiencescore']
        tomatometerstate=BeautifulSoup(get_text,'html.parser').find("score-board")['tomatometerstate']
        tomatometerscore=BeautifulSoup(get_text,'html.parser').find("score-board")['tomatometerscore']

        #Append to list   
        movieinfo.append(audiencescore)
        movieinfo.append(tomatometerscore)
        movieinfo.append(tomatometerstate)
        rowlabel.append('audiencescore')
        rowlabel.append('tomatometerscore')
        rowlabel.append('tomatometerstate')

        df=pd.concat([df,pd.DataFrame(data=movieinfo,columns=[movietitle],index=[rowlabel])],axis=1)
    
    #if URL is incorrect pass
    except Exception as e:
        pass
```

```python
#Run loop and split and rename columns with multiple values
for col in ['Director:','Distributor:','Genre:','Producer:','Writer:']: 
  df_=pd.concat([df_,df_[col].str.split(',', expand=True)],axis=1)
  max_vals=max(df_[col].str.split(',', expand=True).count(axis=1))
  old_columns = list(df_.columns[-max_vals:])
  new_col_list = list(col.lower().replace(':','')+str(num) for num in list(df_.columns[-max_vals:]))
  # rename the columns inplate
  df_.rename(columns={old_columns[idx]: name for  (idx, name) in enumerate(new_col_list)}, inplace=True)
  #drop original column
  df_.drop(columns=[col],inplace=True)
```

Release Date
___

```python
from datetime import datetime

#Extract date only in new release date theater column
df_['release_date_theaters']=df_['Release Date (Theaters):'].apply(lambda x: pd.Series(str(x)[:12]))

#Extra type of release from release date theater column
df_['release_date_theaters_type']=df_['Release Date (Theaters):'].apply(lambda x: pd.Series(''.join(str(x).split())[10:]))

#Convert release date columns to datetime format
df_['Release Date (Streaming):']= pd.to_datetime(df_['Release Date (Streaming):'])
df_['release_date_theaters']= pd.to_datetime(df_['release_date_theaters'])

#rename cell values
df_['release_date_theaters_type']=df_['release_date_theaters_type'].replace({'ide': 'wide', 'imited': 'limited'})

#Create new column for release months (streaming and theaters)
df_['release_month_streaming']=df_['Release Date (Streaming):'].dt.month.astype('Int64')
df_['release_month_theaters']=df_['release_date_theaters'].dt.month.astype('Int64')
#Drop original columns 
df_.drop(columns=['Release Date (Theaters):','Release Date (Streaming):','release_date_theaters'],inplace=True)
```

```python
#Create function to convert box office column values to $M
def KtoM(value):
    value=str(value).replace('$','')
    if value[-1]=='K':
        value=float(value.replace('K',''))/1000
    elif value[-1]=='M':
        value=float(value.replace('M',''))
    return value
```

```python
#Apply function and rename box office column
df_['Box Office (Gross USA):']=df_['Box Office (Gross USA):'].apply(KtoM)
df_ = df_.rename(columns={'Box Office (Gross USA):': 'boxofficegross_inM'})
```
